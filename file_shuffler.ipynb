```python
import pandas as pd
import os
import glob
import shutil
import random
```

```python
random.seed(9) # seed so other people can shuffle it in the same order

image_list = []
label_list = []

for image_path in ["images"]:

    file_list = os.path.join(os.path.join(image_path, "*.*"))
    file_list = glob.glob(file_list)
    previous = len(image_list)
    for name in file_list:
        if ".txt" not in name:
            image_list.append(name)
    print(image_path, len(image_list) - previous)
print(len(image_list))
```

```python
random.shuffle(image_list)
```

```python
mask_list = []
for name in image_list:
    file = os.path.basename(os.path.splitext(name)[0])
    mask_list.append(file + ".png")
```

```python
directory = "./valid/"
check_path = os.path.isdir(directory)
if not check_path:
    os.makedirs(directory)
    print("created folder : ", directory)    
dir_14 = "./train_14_seg/"
check_path = os.path.isdir(dir_14)
if not check_path:
    os.makedirs(dir_14)
    print("created folder : ", dir_14)    
dir_p  = "./train_p_seg/"
check_path = os.path.isdir(dir_p)             
if not check_path:
    os.makedirs(dir_p)
    print("created folder : ", dir_p)   
dir_r = "./train_r_seg/"
check_path = os.path.isdir(dir_r)
if not check_path:
    os.makedirs(dir_r)
    print("created folder : ", dir_r)     
train_count = 0
 
for i in range(0, int(len(image_list)*0.6)+1):
    name = glob.glob(image_list[i]+"*")
    if(len(name) > 0):
        train_count += 1
        if os.path.exists(image_list[i]):
            shutil.copy(image_list[i], directory)
        if os.path.exists(f"./14_masks/{mask_list[i]}"):
            shutil.copy(f"./14_masks/{mask_list[i]}", dir_14)
        if os.path.exists(f"./porcelain_masks/{mask_list[i]}"):
            shutil.copy(f"./porcelain_masks/{mask_list[i]}", dir_p)
        if os.path.exists(f"./recloser_masks/{mask_list[i]}"):
            shutil.copy(f"./recloser_masks/{mask_list[i]}", dir_r)
    else:
        print(image_list[i])

valid_count = 0
element = i
                          
directory = "./valid/"
check_path = os.path.isdir(directory)
if not check_path:
    os.makedirs(directory)
    print("created folder : ", directory)    
dir_14 = "./valid_14_seg/"
check_path = os.path.isdir(dir_14)
if not check_path:
    os.makedirs(dir_14)
    print("created folder : ", dir_14)    
dir_p  = "./valid_p_seg/"
check_path = os.path.isdir(dir_p)             
if not check_path:
    os.makedirs(dir_p)
    print("created folder : ", dir_p)   

dir_r = "./valid_r_seg/"
check_path = os.path.isdir(dir_r)
if not check_path:
    os.makedirs(dir_r)
    print("created folder : ", dir_r)  
                          
for i in range(element, element + int(len(image_list)*0.2)+1):
    name = glob.glob(image_list[i]+"*")
    if(len(name) > 0):
        valid_count += 1
        if os.path.exists(image_list[i]):
            shutil.copy(image_list[i], directory)
        if os.path.exists(f"./14_masks/{mask_list[i]}"):
            shutil.copy(f"./14_masks/{mask_list[i]}", dir_14)
        if os.path.exists(f"./porcelain_masks/{mask_list[i]}"):
            shutil.copy(f"./porcelain_masks/{mask_list[i]}", dir_p)
        if os.path.exists(f"./recloser_masks/{mask_list[i]}"):
            shutil.copy(f"./recloser_masks/{mask_list[i]}", dir_r)
    else:
        print(image_list[i])

test_count = 0
element = i
directory = "./test/"
check_path = os.path.isdir(directory)
if not check_path:
    os.makedirs(directory)
    print("created folder : ", directory)    
dir_14 = "./test_14_seg/"
check_path = os.path.isdir(dir_14)
if not check_path:
    os.makedirs(dir_14)
    print("created folder : ", dir_14)    
dir_p  = "./test_p_seg/"
check_path = os.path.isdir(dir_p)
if not check_path:
    os.makedirs(dir_p)
    print("created folder : ", dir_p)   
dir_r = "./test_r_seg/"
check_path = os.path.isdir(dir_r)
if not check_path:
    os.makedirs(dir_r)
    print("created folder : ", dir_r)  

                          
for i in range(element, element + int(len(image_list)*0.2)):
    name = glob.glob(image_list[i]+"*")
    if(len(name) > 0):
        test_count += 1
        if os.path.exists(image_list[i]):
            shutil.copy(image_list[i], directory)
        if os.path.exists(f"./14_masks/{mask_list[i]}"):
            shutil.copy(f"./14_masks/{mask_list[i]}", dir_14)
        if os.path.exists(f"./porcelain_masks/{mask_list[i]}"):
            shutil.copy(f"./porcelain_masks/{mask_list[i]}", dir_p)
        if os.path.exists(f"./recloser_masks/{mask_list[i]}"):
            shutil.copy(f"./recloser_masks/{mask_list[i]}", dir_r)
    else:
        print(image_list[i])

print(f'Total: {train_count + valid_count + test_count}')
print(f'Train: {train_count}, Valid: {valid_count}, Test: {test_count}')
print(f'Total: {train_count + valid_count}')
print(f'Train: {train_count}, Valid: {valid_count}')
```
